```python
plt.figure(figsize=(10, 6))
plt.bar(df['country'], df['yearly-average-temp'])
plt.xlabel('Country')
plt.ylabel('Temperature (°C)')
plt.title('Temperature by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['yearly-average-temp'], bins=10, edgecolor='black')
plt.xlabel('Country')
plt.ylabel('Temperature (°C)')
plt.title('Temperature by Country')
plt.xticks(rotation=45)
plt.show()
```
